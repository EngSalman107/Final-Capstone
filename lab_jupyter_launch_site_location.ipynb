```python
# TASK 3: Calculate the distances between a launch site to its proximities
import numpy as np
# : Define the Haversine function to calculate the distance
def haversine(lat1, long1, lat2, long2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, long1, lat2, long2])

# Calculate differences
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    
    # Haversine formula
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    
    # Radius of Earth in kilometers
    r = 6371
    return c * r

# : Calculate distances between all launch sites
# Create a distance matrix
distance_matrix = pd.DataFrame(index=launch_sites_df['Launch Site'], columns=launch_sites_df['Launch Site'])

for i in range(len(launch_sites_df)):
    for j in range(len(launch_sites_df)):
        if i != j:  # Avoid calculating distance from a site to itself
            lat1 = launch_sites_df.iloc[i]['Lat']
            lon1 = launch_sites_df.iloc[i]['Long']
            lat2 = launch_sites_df.iloc[j]['Lat']
            lon2 = launch_sites_df.iloc[j]['Long']
            
            distance = haversine(lat1, lon1, lat2, lon2)
            distance_matrix.iloc[i, j] = distance

# : Display the distance matrix
distance_matrix
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
